```python
from langchain_ollama import ChatOllama
llm = ChatOllama(model="qwen2.5-coder:7b", base_url="http://localhost:11434")
```

# **First Graph**

### **Step 1: Define the Schema**

```python
from typing import TypedDict,List
class graph_schema(TypedDict):
  name : str
  message: str
```

### **Step2: Create the Node Function**

```python
def welcome(state:graph_schema):
  current_name = state['name']
  current_message = state['message']
  
  respone = llm.invoke(f'My name is {current_name}.{current_message}').content
  state['message'] = f"AssertionErrorHere is your message : = {current_message}. Here is my response{respone}"
  
  return state
    
```

### **Step3 : Create the Graph**

```python
from langgraph.graph import StateGraph, MessagesState, START, END



graph = StateGraph(graph_schema)
graph.add_node(welcome)
graph.add_edge(START, "welcome")
graph.add_edge("welcome", END)
graph = graph.compile()


```

```python
#graph.invoke({"messages": [{"role": "user", "content": "hi!"}]})
response = graph.invoke({"name": "John", "message": "Tell me a joke"})
print(response)

```
